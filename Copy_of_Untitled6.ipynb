```python
import pandas as pd
import io

csv_data = """
Month,Product,Units_Sold,Unit_Price,Total_Sales
Jan,Laptop,10,650,6500
Jan,Phone,25,200,5000
Jan,Headphones,40,25,1000
Feb,Laptop,12,650,7800
Feb,Phone,30,200,6000
Feb,Headphones,38,25,950
Mar,Laptop,9,650,5850
Mar,Phone,28,200,5600
Mar,Headphones,50,25,1250
Apr,Laptop,15,650,9750
Apr,Phone,35,200,7000
Apr,Headphones,45,25,1125
May,Laptop,14,650,9100
May,Phone,40,200,8000
May,Headphones,55,25,1375
"""

df = pd.read_csv(io.StringIO(csv_data))
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# The dataset is already loaded into 'df' from a previous cell.
# No need to load it again from 'sales_data.csv'.

# Show first rows
print("FIRST 5 ROWS:\n")
print(df.head())

# Summary statistics
print("\nSUMMARY STATISTICS:\n")
print(df.describe())

# Total sales per month
print("\nTOTAL SALES PER MONTH:\n")
print(df.groupby("Month")["Total_Sales"].sum())

# Total sales per product
print("\nTOTAL SALES PER PRODUCT:\n")
print(df.groupby("Product")["Total_Sales"].sum())

# Bar chart - monthly sales
df.groupby("Month")["Total_Sales"].sum().plot(kind="bar")
plt.title("Monthly Total Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# Pie chart - product sales distribution
df.groupby("Product")["Total_Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales by Product")
plt.ylabel("")  # Removes the weird pie chart label
plt.show()
```
